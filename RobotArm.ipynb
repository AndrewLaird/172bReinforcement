```python
import gym
import numpy as np
```

```python
env = gym.make("FetchPickAndPlace-v1")
```

```python
print(env.action_space)
```

```python
env.reset()
for i in range(1000):
    action = env.action_space.sample()
    #print(action)
    env.render()
    new_state, reward, is_done, _= env.step(action)
    print(new_state)
```
